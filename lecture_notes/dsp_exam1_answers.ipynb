## Part 3.

Determine whether the resulting linear transformations are linearly dependent. Provide necessary solutions

Let
$$
A = \begin{pmatrix}
5 & 0 & 0 \\
0 & 5 & 0 \\
0 & 0 & 5
\end{pmatrix}
$$
$$
B = \begin{pmatrix}
1 & 0 & 5 \\
2 & 7 & 6 \\
6 & 4 & 7
\end{pmatrix}
$$
$$X = A \cdot B$$

$
X = 5 \begin{pmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{pmatrix}
\cdot \begin{pmatrix}
1 & 0 & 5 \\
2 & 7 & 6 \\
6 & 4 & 7
\end{pmatrix} = 5I \cdot \begin{pmatrix}
1 & 0 & 5 \\
2 & 7 & 6 \\
6 & 4 & 7
\end{pmatrix} = 5\begin{pmatrix}
1 & 0 & 5 \\
2 & 7 & 6 \\
6 & 4 & 7
\end{pmatrix} = \begin{pmatrix}
5 & 0 & 25 \\
10 & 35 & 30 \\
30 & 20 & 35
\end{pmatrix} 
$

$|X| = 5\left| \begin{pmatrix}
35 & 30 \\
20 & 35
\end{pmatrix}\right| - 0\left| \begin{pmatrix}
10 & 30 \\
30 & 35
\end{pmatrix}\right| + 25\left| \begin{pmatrix}
10 & 35 \\
30 & 20
\end{pmatrix}\right| = 5\left| \begin{pmatrix}
35 & 30 \\
20 & 35
\end{pmatrix}\right| + 25\left| \begin{pmatrix}
10 & 35 \\
30 & 20
\end{pmatrix}\right|$

$|X| = 5 \left[ (35 \cdot 35) - (30 \cdot 20) \right] + 25 \left[ (10 \cdot 20) - (30 \cdot 35) \right]$

$|X| = 5 \left[ 1225 - 600 \right] + 25 \left[200 - 1050 \right]$

$|X| = 5\cdot 625 + 25 \left(-850 \right)$

$|X| = 3125 - 21250 = -18125$

$\therefore X \text{ is linearly independent}$

Let
$$
C = \begin{pmatrix}

\end{pmatrix}
$$
$$
D = \begin{pmatrix}

\end{pmatrix}
$$
$$M = C \cdot D$$

$
M = \begin{pmatrix}
1 & 2 & 6 \\
3 & 15 & 4 \\
2 & 10 & 3
\end{pmatrix}
\cdot \begin{pmatrix}
5 & 2 & 4 \\
6 & 2 & 4 \\
0 & 1 & 1
\end{pmatrix} = \begin{pmatrix}
17 &  12 &  18 \\
105&  40 &  76 \\
70&  27&  51
\end{pmatrix} 
$

$|M| = 17\left| \begin{pmatrix}
40 & 76 \\
27 & 51
\end{pmatrix}\right| + 12\left| \begin{pmatrix}
105 & 76 \\
70 & 51
\end{pmatrix}\right| - 18\left| \begin{pmatrix}
105 & 40 \\
76 & 27
\end{pmatrix}\right|
$

$|M| = 17 \left[ (40 \cdot 51) - (76 \cdot 27) \right] + 12 \left[ (105 \cdot 51) - (76 \cdot 70) \right] - 18 \left[ (105 \cdot 27) - (70 \cdot 40) \right]$

$|M| = 17 \left[ 2040 - 2052 \right] + 12 \left[ 5355 - 5320 \right] - 18 \left[2835 - 2800 \right]$

$|M| = 17(-12) + 12(35) - 18(35)$

$|M| = -204 - 420 + 630 = 6$

$\therefore M \text{ is linearly independent}$

```python
C = np.array([
    [1,2,6],
    [3,15,4],
    [2,10,3]
])
D = np.array([
    [5,2,4],
    [6,2,4],
    [0,1,1]
])
M = C@D
M
```

## Part 4.

Plot the signal corresponding to the following vectors whereas $T$ is the time vector and $G$ is the vector corresponding to the amplitudes of the signal.

$$T = \begin{pmatrix} 0 & \frac{\pi}{4} & \frac{\pi}{2} & \frac{3\pi}{4} & \pi\end{pmatrix}$$
$$G = \begin{pmatrix} 5 & 3 & 0 & -3 & 5\end{pmatrix}$$

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter, MultipleLocator
```

```python
T = np.array([0, np.pi/4, np.pi/2, 3*np.pi/4, np.pi])
G = np.array([5, 3, 0, -3, 5])
```

```python
fig, ax = plt.subplots(figsize=(7,3))

ax.plot(T,G, 'bo:')
ax.stem(T,G)

ax.set_ylabel(r'Amplitude [$G$]')
ax.set_xlabel(r'Time [$T$]')

ax.xaxis.set_major_formatter(FuncFormatter(
   lambda val,pos: '{:.2f}$\pi$'.format(val/np.pi) if val !=0 else '0'
))
ax.xaxis.set_major_locator(MultipleLocator(base=np.pi/4))
plt.grid()
```

## Bonus Graphs for Part 1

```python
def plot_complexmat(mat, title):

    vect1_re = mat[:,0].real
    vect2_re = mat[:,1].real
    
    vect1_im = mat[:,0].imag
    vect2_im = mat[:,1].imag
    
    fig, axes = plt.subplots(1,2, figsize=(10,4))
    
    
    fig.suptitle(f'Vector plot for {title}')
    axes[0].set_title('Real Plot')
    axes[0].quiver(0,0, vect1_re[0]+1e-2, vect1_re[1]+1e-2, label='vector 1', color='red',
                   scale=1, scale_units='xy', angles='xy')
    axes[0].quiver(0,0, vect2_re[0]+1e-2, vect2_re[1]+1e-2, label='vector 2', color='blue',
                   scale=1, scale_units='xy', angles='xy')
    
    axes[1].set_title('Imaginary Plot')
    axes[1].quiver(0,0, vect1_im[0]+1e-2, vect1_im[1]+1e-2, label='vector 1', color='red',
                   scale=1, scale_units='xy', angles='xy')
    axes[1].quiver(0,0, vect2_im[0]+1e-2, vect2_im[1]+1e-2, label='vector 2', color='blue',
                   scale=1, scale_units='xy', angles='xy')
    
    for i in range(len(axes)):
        axes[i].set_xlim(-2,2)
        axes[i].set_ylim(-2,2)
        axes[i].legend()
        axes[i].grid()
```

```python
## Part 1.1 HY
plot_complexmat(manual_1_1, r'$H\cdot Y$')
```

```python
## Part 1.2 YH
plot_complexmat(manual_1_2, r'$Y\cdot H$')
```

```python
## Part 1.3 HH
plot_complexmat(manual_1_3, r'$H\cdot H$')
```

```python
## Part 1.4 YHO
plot_complexmat(manual_1_4, r'$Y \cdot H \cdot O$')
```

```python
## Part 1.5 HYHO
plot_complexmat(manual_1_5, r'$H \cdot Y \cdot H \cdot O$')
```
